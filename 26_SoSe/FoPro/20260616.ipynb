```r
library(tidyverse)
```

Pseudo-Daten erstellen, um Zeitdauer zu berechnen:

```r
d_raw <-
  tibble(
    start_time = c("2026-06-16 08:00:00", "2026-06-16 08:01:00"),
    end_time = c("2026-06-16 09:00:00", "2026-06-16 08:02:00")
  )
```

```r
d_raw |>
  mutate(
    start_time      = ymd_hms(start_time),
    end_time        = ymd_hms(end_time),
    dauer           = end_time - start_time,
    dauer_kurz      = dauer < dminutes(5)
  )
```
